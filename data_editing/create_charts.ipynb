```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Daten anpassen und die Technologie-Branche entfernen
data = {
    "Kategorie": ["Sportbranche", "Sportbranche", "Progressive Marken", "Progressive Marken",
                  "Lebensmittel", "Lebensmittel", "Transport & Logistik", "Transport & Logistik",
                  "Regressive Marken", "Regressive Marken"],
    "Brand Activism": ["Mit Brand Activism", "Ohne Brand Activism", "Mit Brand Activism", "Ohne Brand Activism",
               "Mit Brand Activism", "Ohne Brand Activism", "Mit Brand Activism", "Ohne Brand Activism",
               "Mit Brand Activism", "Ohne Brand Activism"],
    "Likes": [15112, 26079, 19372, 3536, 2426, 4495, 1721, 2626, 874, 2625],
    "Kommentare": [177, 234, 882, 90, 34, 145, 145, 141, 253, 613]
}

df = pd.DataFrame(data)
plt.rc('font', size=12)
sns.set(font_scale=1.6)
# Grafiken erstellen (ohne die Technologie-Branche)
plt.figure(figsize=(14, 8))

sns.barplot(x='Kategorie', y='Likes', hue='Brand Activism', data=df, palette="grey")
plt.title('Anzahl der Likes nach Kategorie und Brand Activism')
plt.ylabel('Likes')
plt.xlabel('Kategorie')
plt.legend(title='Brand Activism')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

sns.barplot(x='Kategorie', y='Kommentare', hue='Brand Activism', data=df, palette="grey")
plt.title('Anzahl der Kommentare nach Kategorie und Brand Activism')
plt.ylabel('Kommentare')
plt.xlabel('Kategorie')
plt.legend(title='Brand Activism')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

# Aktualisiertes DataFrame fr Prozentuale Unterschiede ohne Technologie-Branche
difference_data = {
    "Kategorie": ["sports_industry", "Progressive Marken",
                  "Lebensmittel", "Transport & Logistik", "regressive_brands"],
    "Likes Unterschied in %": [72.57, 547.89, 85.26, 52.53, 200.29],
    "Kommentare Unterschied in %": [32.20, 985.47, 329.26, -2.77, 142.80]
}

df_diff = pd.DataFrame(difference_data)

# Grafische Darstellung der Prozentualen Unterschiede ohne Technologie-Branche
fig, ax1 = plt.subplots(figsize=(14, 8))

sns.barplot(x='Kategorie', y='Likes Unterschied in %', data=df_diff, color='black', ax=ax1)
ax1.set_ylabel('Likes Unterschied in %', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xlabel('Kategorie')

ax2 = ax1.twinx()
sns.lineplot(x='Kategorie', y='Kommentare Unterschied in %', data=df_diff, color='grey', marker='o', ax=ax2)
ax2.set_ylabel('Kommentare Unterschied in %', color='grey')
ax2.tick_params(axis='y', labelcolor='grey')

plt.title('Prozentualer Unterschied zwischen Brand Activism und Ohne Brand Activism')
plt.xticks(rotation=45)
fig.tight_layout()

plt.show()
```
